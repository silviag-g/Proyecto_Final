```python
#Training set

target = 'price'

train_dataframe = pd.get_dummies(train)
train_dataframe
```

```python
test_df = pd.get_dummies(test)
test_df
```

```python
train_df = train_dataframe.drop('price', axis=1)
train_df = train_df.reset_index().rename(columns={'index':'id'})
train_df
```

```python
target_df = train_dataframe[['price']]
target_df
```

```python
#Scaler train
scaler = StandardScaler()
scaled_train = scaler.fit_transform(train_df)
scaled_train
```

```python
#Scaler train
scaler = StandardScaler()
scaled_test = scaler.fit_transform(test_df)
scaled_test
```

```python
model = LinearRegression()
model.fit(scaled_train, target_df)
```

```python
X_test= scaler.fit_transform(scaled_test)
y_hat = model.predict(X_test).clip(0,30000) #esta es la predicciÃ³n. es un array de numpy
```
